```python
class Solution:
    def numUniqueEmails(self, emails: 'List[str]') -> 'int':
        vocabulary_count = {}
        for i in emails:
            local, domain = i.split('@')
            local = local.partition('+')[0]
            local = local.replace('.', '')
            local = local + '@' + domain
            if local not in vocabulary_count:
                vocabulary_count[local] = emails.index(i) 
        return len(vocabulary_count)
    
    def numJewelsInStones(self, J: 'str', S: 'str') -> 'int':
        counter = 0
        for letter in J:
            counter += S.count(str(letter))
        return counter

    def uniqueMorseRepresentations(self, words: 'List[str]') -> 'int':
        seen = set()
        combination = ''
        table = [".-","-...","-.-.","-..",".","..-.","--.","....","..",
                 ".---","-.-",".-..","--","-.","---",".--.","--.-",".-.",
                 "...","-","..-","...-",".--","-..-","-.--","--.."]
        for word in words:
            combination = ''.join(table[ord(str(letter)) - 97] for letter in word)
            seen.add(combination)
        return len(seen)
    
    def flipAndInvertImage(self, A: 'List[List[int]]') -> 'List[List[int]]':
        for node in A:
            left, right = 0 , len(A) - 1
            while left <= right:
                if node[left] == node[right]:
                    node[left], node[right] = 1 - node[left], 1 - node[right]
                    left += 1
                    right -= 1
                else:
                    left += 1
                    right -= 1
        return A 

    def judgeCircle(self, moves: 'str') -> 'bool':
        gorizontal, vertical = 0, 0
        for letter in moves:
            if letter == 'U':
                vertical += 1
            if letter == 'D':
                vertical -= 1
            if letter == 'R':
                vertical += 1
            if letter == 'L':
                vertical -= 1
        return (gorizontal, vertical) == (0, 0)
        
    def diStringMatch(self, S: 'str') -> 'List[int]': 
        ans = []
        numbers = list(range(len(S) + 1))
        for number in range(len(S)):
            if S[number] == 'I' and number == 0:
                ans.append(numbers.pop(0))
            elif S[number] == 'D' and number == 0:
                ans.append(numbers.pop())
            elif S[(number - 1):(number + 1)] == 'II':
                ans.append(numbers.pop(0))
            elif S[(number - 1):(number + 1)] == 'ID':
                ans.append(numbers.pop())
            elif S[(number - 1):(number + 1)] == 'DI':
                ans.append(numbers.pop(0))
            elif S[(number - 1):(number + 1)] == 'DD':
                ans.append(numbers.pop())
        ans.append(numbers.pop())  
        return ans
    
    def minDeletionSize(self, A: 'List[str]') -> 'int':
        counter = 0
        for string in range(len(A) - 1):
            for letter in range(len(A[0])):
                if A[string][letter] > A[string + 1][letter]:
                    counter += 1
        return counter

    def selfDividingNumbers(self, left: 'int', right: 'int') -> 'List[int]':
        ans = []
        
        def isNumberSelf(number):
            numbers = set(int(x) for x in str(number))
            if 0 in numbers:
                pass
            elif all(map(lambda x: number % x == 0, numbers)):
                return number
        
        for cifar in range(left, right + 1):
            if isNumberSelf(cifar):
                ans.append(isNumberSelf(cifar))
        
        return ans
    
    def sumEvenAfterQueries(self, A: 'List[int]', queries: 'List[List[int]]') -> 'List[int]':
        
        ans = []
        
        for query in queries:
            
            previous = A[query[1]]
            A[query[1]] += query[0]
            
            if not ans:
                ans.append(sum([x for x in A if x % 2 == 0]))
            
            elif A[query[1]] % 2 != 0 and previous % 2 == 0:
                ans.append(ans[-1] - previous)
            
            elif A[query[1]] % 2 != 0 and previous % 2 != 0:
                ans.append(ans[-1])
                
            elif A[query[1]] % 2 == 0 and previous % 2 != 0:
                ans.append(ans[-1] + A[query[1]])
                
            else:
                ans.append(ans[-1] + A[query[1]] - previous)
        return ans
    
    def commonChars(self, A: 'List[str]') -> 'List[str]':
        letters = []
        for small_letter in set(A[0]):
            count = min([word.count(small_letter) for word in A])
            if count is not 0:
                letters.extend(small_letter * count)
        return letters
    
    
a = Solution()
```
